```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


with open("data2.json") as fp:
    data = json.load(fp)
    
print(data.keys())
```

```python
flat_data = []
for ts, vehicles in zip(data["ts"], data["vehiclesU"]):
    for vehicle in vehicles:
        flat_data.append({"id": vehicle["id"], "u": vehicle["u"], "ts": ts})

# Create a DataFrame from the flattened data
df = pd.DataFrame(flat_data)
print(df)
```

```python
# lets look only at the first 50
plt.plot(data['vehCount'][51:100])
print(f"flows: {data['flows'][101]}, speeds: {data['speeds'][101]}, timeframe: {data['ts'][51]} - {data['ts'][100]}")
```

```python
def calc_params(data, df, start, stop):
    curr_count = 0
    counts = []
    u = data["u"]
    past_leaders = set()
    speed_sum = 0

    ndf = df.copy(deep=True)
    ndf["dt"] = df.groupby("id")["ts"].diff()
    ndf["speed"] = df.groupby("id")["u"].diff() / ndf["dt"]

    for i, ts in enumerate(data['ts'][start:stop]):
        # is sorted by u position, in descending order
        vehs = ndf[np.isclose(df["ts"].values, ts)][["id", "u", "speed"]]
        vehs = vehs[vehs["u"] > u]
        # get what they call the 'leader', that is the last car that passed by u
        leader = vehs["id"].values[-1]
        if i == 0:
            past_leaders.add(leader)
            counts.append(curr_count)
            continue
        if leader not in past_leaders:
            past_leaders.add(leader)
            speed_sum += vehs["speed"].values[-1]
            curr_count += 1
        counts.append(curr_count)
    # in the js code flow is calculated as `3600*this.vehCount/this.dtAggr`
    flow = 3600 * curr_count / data["dtAggr"]
    # in the js code flow is calculated as `Math.round(3.6*(this.speedSum/this.vehCount))`
    speed = np.round(3.6 * speed_sum / curr_count)
    # in the js code density is calculated as 1000 * flow / speed
    dens = 1000 * flow / speed
    return flow, speed, dens, counts


def calc_params2(data, df, start, stop):
    curr_count = 0
    counts = []
    u = data["u"]
    past_leaders = set()
    speed_sum = 0

    ndf = df.copy(deep=True)
    ndf["dt"] = df.groupby("id")["ts"].diff()
    ndf["speed"] = df.groupby("id")["u"].diff() / ndf["dt"]

    for i, ts in enumerate(data['ts'][start:stop]):
        # is sorted by u position, in descending order
        vehs = ndf[np.isclose(df["ts"].values, ts)][["id", "u", "speed"]]
        vehs = vehs[vehs["u"] >= u]
        # get what they call the 'leader', that is the last car that passed by u
        leaders = vehs["id"]
        if i == 0:
            for leader in leaders:
                past_leaders.add(leader)
            counts.append(curr_count)
            continue
        for leader in leaders:
            if leader in past_leaders:
                continue
        
            past_leaders.add(leader)
            speed_sum += vehs[vehs["id"]==leaders]["speed"].values[0]
            curr_count += 1
        counts.append(curr_count)
    # in the js code flow is calculated as `3600*this.vehCount/this.dtAggr`
    flow = 3600 * curr_count / data["dtAggr"]
    # in the js code flow is calculated as `Math.round(3.6*(this.speedSum/this.vehCount))`
    speed = np.round(3.6 * speed_sum / curr_count)
    # in the js code density is calculated as 1000 * flow / speed
    dens = 1000 * flow / speed
    return flow, speed, dens, counts
```

```python
last = len(data['vehCount'])
start = 0
i = 0
stop = 0
while True:
    i += 1
    if i == last:
        flow, speed, dens, counts = calc_params(data, df, start, stop+1)
        print(f"flow error at i={start}: {np.abs(flow - data['flows'][stop])}")
        print(f"speed error at range={start}-{stop}: {np.abs(speed - data['speeds'][stop])}")
        break
    if data['vehCount'][stop] > data['vehCount'][i]:
        flow, speed, dens, counts = calc_params(data, df, start, stop+1)
        print(f"flow error at range={start}-{stop}: {np.abs(flow - data['flows'][stop+1])}")
        print(f"speed error at range={start}-{stop}: {np.abs(speed - data['speeds'][stop+1])}")
        print()
        plt.show()
        start = i
    stop += 1
    
```

```python
start = 349
stop = start + 50
flow, speed, dens, counts = calc_params(data, df, start, stop)
print(flow, data['flows'][stop+1], counts[-1])
plt.plot(counts, "o-")
plt.plot(data['vehCount'][start:stop], "o")
plt.show()
plt.plot(data['flows'][start:stop+1], "o-")
```
